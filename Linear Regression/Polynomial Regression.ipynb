```python
# Importing Boston Dataset
from sklearn import datasets
data = datasets.load_boston()
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
X = data.data
Y = data.target
```

```python
df = pd.DataFrame(X)
df.columns = data.feature_names
df.describe()
```

```python
df["CRIM_CRIM"] = df.CRIM ** 2
df["ZN_ZN"] = df.ZN ** 2
df["INDUS_INDUS"] = df.INDUS ** 2
df["CHAS_CHAS"] = df.CHAS ** 2
df["NOX_NOX"] = df.NOX ** 2
df["RM_RM"] = df.RM ** 2
df["AGE_AGE"] = df.AGE ** 2
df["DIS_DIS"] = df.DIS ** 2
df["RAD_RAD"] = df.RAD ** 2
df["TAX_TAX"] = df.TAX ** 2
df["PTRATIO_PTRATIO"] = df.PTRATIO ** 2
df["B_B"] = df.B ** 2
df["LSTAT_LSTAT"] = df.LSTAT ** 2
```

```python
# Now comes the training part
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
from sklearn import model_selection
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, random_state = 20)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(df.values, Y, random_state = 20)
```

```python
algo1 = LinearRegression()
algo2 = LinearRegression()
```

```python
algo1.fit(X_train, Y_train)
algo2.fit(X2_train, Y2_train)
```

```python
# Test Scores

# Algo 1 Scores
print("Train Score Algo 1 : ", algo1.score(X_train, Y_train))
print("Test Score Algo 1 : ", algo1.score(X_test, Y_test))

print()

# Algo 2 scores
print("Train score Algo 2 : ", algo2.score(X2_train, Y2_train))
print("Test score Algo 2 : ", algo2.score(X2_test, Y2_test))
```
